```python
import pandas as pd

df = pd.read_csv("AI_Resume_Screening.csv")
df.head()
```

```python
print(df.columns)
```

```python
import random
df['Gender'] = [random.choice(['Male', 'Female']) for _ in range(len(df))]
df['College Tier'] = [random.choice(['Tier-1', 'Tier-2', 'Tier-3']) for _ in range(len(df))]
```

```python
df[['Name', 'Gender', 'College Tier']].head()
```

```python
df = df.drop(columns=['Name', 'Gender', 'College Tier'], errors='ignore')

df.head()
```

```python
print(df.columns)
```

```python
# Count number of skills
df['skill_count'] = df['Skills'].apply(lambda x: len(str(x).split(',')))

# Experience score
df['experience_score'] = df['Experience (Years)']

# Count certifications
df['cert_count'] = df['Certifications'].apply(lambda x: len(str(x).split(',')))

# Education score mapping
edu_map = {
    'b.tech': 3,
    'mba': 4,
    'phd': 5
}

def education_score(x):
    return edu_map.get(str(x).lower(), 1)

df['education_score'] = df['Education'].apply(education_score)

# Project score
df['project_score'] = df['Projects Count']

df.head()
```

```python
df['domain_match'] = df['Job Role']
```

```python
df.to_csv("processed_resume_data.csv", index=False)
```
